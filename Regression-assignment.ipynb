## Q1. What is Simple Linear Regression?
Simple Linear Regression is a statistical technique used to model the relationship between one independent variable (X) and one dependent variable (Y) using a straight line equation: Y = β0 + β1X.

## Q2. What are the key assumptions of Simple Linear Regression?
The key assumptions are: linear relationship between X and Y, independence of errors, homoscedasticity (constant variance of errors), normal distribution of errors, and no multicollinearity.

## Q3. What is heteroscedasticity, and why is it important to address in regression models?
Heteroscedasticity occurs when the variance of residuals changes across values of the independent variable. It is important to address because it can make coefficient estimates unreliable and affect hypothesis testing.

## Q4. What is Multiple Linear Regression?
Multiple Linear Regression models the relationship between one dependent variable and two or more independent variables. The equation is Y = β0 + β1X1 + β2X2 + ... + βnXn.

## Q5. What is polynomial regression, and how does it differ from linear regression?
Polynomial regression models a non-linear relationship by adding polynomial terms (X², X³, etc.). Unlike linear regression, it can fit curved relationships.

## Q6. Simple Linear Regression with Plot

```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

model = LinearRegression()
model.fit(X, Y)

Y_pred = model.predict(X)

plt.scatter(X, Y)
plt.plot(X, Y_pred, color='red')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simple Linear Regression")
plt.show()
```

## Q7. Multiple Linear Regression

```python

import numpy as np
from sklearn.linear_model import LinearRegression

Area = np.array([1200, 1500, 1800, 2000])
Rooms = np.array([2, 3, 3, 4])
Price = np.array([250000, 300000, 320000, 370000])

X = np.column_stack((Area, Rooms))

model = LinearRegression()
model.fit(X, Price)

print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
```

## Q8. Polynomial Regression (2nd Degree)

```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, Y)

X_line = np.linspace(1, 5, 100).reshape(-1, 1)
X_line_poly = poly.transform(X_line)
Y_line = model.predict(X_line_poly)

plt.scatter(X, Y)
plt.plot(X_line, Y_line, color='red')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Polynomial Regression (Degree 2)")
plt.show()
```

## Q9. Residuals Plot to Assess Heteroscedasticity

```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

model = LinearRegression()
model.fit(X, Y)

Y_pred = model.predict(X)
residuals = Y - Y_pred

plt.scatter(X, residuals)
plt.axhline(y=0, color='red')
plt.xlabel("X")
plt.ylabel("Residuals")
plt.title("Residuals Plot")
plt.show()
```
