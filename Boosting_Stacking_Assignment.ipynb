## 1. What is Boosting in Machine Learning?
Boosting is an ensemble technique where weak learners are trained sequentially, with each model focusing more on previous errors to improve accuracy and reduce bias.

## 2. Difference between AdaBoost and Gradient Boosting
AdaBoost adjusts sample weights after each iteration, while Gradient Boosting optimizes a loss function using gradient descent.

## 3. How does regularization help in XGBoost?
Regularization penalizes complex trees in XGBoost using L1 and L2 terms, reducing overfitting.

## 4. Why is CatBoost efficient for categorical data?
CatBoost automatically encodes categorical features and avoids data leakage.

## 5. Applications where boosting is preferred
Boosting is used in credit scoring, fraud detection, medical diagnosis, and churn prediction.

## 6. AdaBoost Classifier on Breast Cancer Dataset

```python
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = AdaBoostClassifier(random_state=42)
model.fit(X_train, y_train)

pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, pred))
```

## 7. Gradient Boosting Regressor on California Housing Dataset

```python
from sklearn.datasets import fetch_california_housing
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

X, y = fetch_california_housing(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = GradientBoostingRegressor(random_state=42)
model.fit(X_train, y_train)

pred = model.predict(X_test)
print("R2 Score:", r2_score(y_test, pred))
```

## 8. XGBoost Classifier with Learning Rate Tuning

```python
from xgboost import XGBClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score

X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

param_grid = {'learning_rate': [0.01, 0.1, 0.2]}
xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss')

grid = GridSearchCV(xgb, param_grid, cv=3)
grid.fit(X_train, y_train)

best = grid.best_estimator_
print("Best Params:", grid.best_params_)
print("Accuracy:", accuracy_score(y_test, best.predict(X_test)))
```

## 9. CatBoost Classifier with Confusion Matrix

```python
from catboost import CatBoostClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = CatBoostClassifier(verbose=0)
model.fit(X_train, y_train)

pred = model.predict(X_test)

cm = confusion_matrix(y_test, pred)
sns.heatmap(cm, annot=True, fmt='d')
plt.show()
```
