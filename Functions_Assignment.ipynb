# Assignment: Functions

### Q1. What is the difference between a function and a method in Python?
A function is a block of code that performs a specific task and can be called directly. A method is a function that is associated with an object or class. Functions are called using their name, while methods are called using an object. Both help in code reuse.

**Example:** `print()` is a function, while `text.upper()` is a method.

### Q2. Explain the concept of function arguments and parameters in Python.
Parameters are variables written in the function definition. Arguments are the actual values passed when calling the function.

**Example:**
`def add(a,b):`
Here `a` and `b` are parameters.
`add(5,3)` → `5` and `3` are arguments.

### Q3. Different ways to define and call a function.
Functions can be defined with no arguments, with arguments, with default arguments, and with keyword arguments.

**Example:** `def greet(name='User'):`

### Q4. Purpose of return statement.
The `return` statement sends a value back from a function. It allows the result to be stored or used later.

**Example:** `return a+b`

### Q5. Iterators and Iterables.
An iterable is an object that can be looped over. An iterator gives one value at a time using `next()`.

**Example:** A list is iterable, while `iter(list)` returns an iterator.

### Q6. Generators.
Generators produce values one by one using the `yield` keyword. They save memory because they do not store all values at once.

**Example:** `yield i`

### Q7. Advantages of generators.
Generators use less memory, improve performance for large data, generate values when needed, and are useful for loops.

**Example:** Reading a large file line by line.

### Q8. Lambda function.
A lambda function is a small anonymous function written in one line. It is commonly used with `map()`, `filter()`, and `sorted()`.

**Example:** `lambda x: x*x`

### Q9. Purpose of map().
`map()` applies a function to every item in an iterable and returns the result.

**Example:** `map(lambda x:x*2,[1,2,3])`

### Q10. Difference between map(), filter(), and reduce().
`map()` changes every element. `filter()` selects elements based on a condition. `reduce()` combines all elements into a single value.

**Example:** `map(x*2)`, `filter(even)`, `reduce(sum)`

```python
# Q1
def even_sum(data):
    total = 0
    for i in data:
        if i % 2 == 0:
            total += i
    return total

print(even_sum([1,2,3,4,5,6]))
```

```python
# Q2
def reverse_string(text):
    return text[::-1]

print(reverse_string("Python"))
```

```python
# Q3
def square_list(data):
    result = []
    for i in data:
        result.append(i*i)
    return result

print(square_list([1,2,3,4]))
```

```python
# Q4
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

for i in range(1,201):
    if is_prime(i):
        print(i,end=" ")
```

```python
# Q5
class Fibonacci:
    def __init__(self,n):
        self.n=n
        self.a=0
        self.b=1
        self.count=0
    def __iter__(self):
        return self
    def __next__(self):
        if self.count>=self.n:
            raise StopIteration
        value=self.a
        self.a,self.b=self.b,self.a+self.b
        self.count+=1
        return value

for i in Fibonacci(10):
    print(i,end=" ")
```

```python
# Q6
def powers(exp):
    for i in range(exp+1):
        yield 2**i

for i in powers(5):
    print(i)
```

```python
# Q7

def read_file(file_name):
    with open(file_name, "r") as file:
        for line in file:
            yield line.strip()

# Calling the generator
for line in read_file("sample.txt"):
    print(line)
```
